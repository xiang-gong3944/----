## 銅酸化物様結晶と交代磁性

## 考える系の状況

### ハミルトニアン

交代磁性についての Smejkal-Sinova の Phys.Rev.X 12040501 にある Ru2 の模式図を見ると、
銅酸化物の二次元正方格子を歪ませたように見える。
なので Toy model として銅酸化物をひずませた模型を作って、
交代磁性が出るかを確認してみるというアイデアが出てもおかしくはない。
しかし、遠山先生曰くなぜかやった結果というのが出回っていないらしい。

なのでこのモデルを調べて、相図やスピン伝導度、超伝導との関連も見たいといった目標である。

一般的な銅酸化物をモデルとした模型のハミルトニアンは次のように書ける。
$$
\begin{align*}
    \mathcal{H} 
        &= \mathcal{H}_\varepsilon + \mathcal{H}_U + \mathcal{H}_T\\
    \mathcal{H}_\varepsilon 
        &= \varepsilon_d \sum_{i,\sigma} n_{i,\sigma}^d + (\varepsilon_d +\Delta)\sum_{i,\delta,\sigma}n_{i+\delta/2,\sigma}^p \\
    \mathcal{H}_U 
        &= U_d \sum_i n_{i,\uparrow}^d n_{i,\downarrow}^d 
            + U_p \sum_{i,\delta}n_{i+\delta/2,\uparrow}^p n_{i+\delta/2,\downarrow}^p
            + U_{pd} \sum_{i,\delta,\sigma,\sigma'}n_{i,\sigma}^p n_{i+\delta/2,\sigma'}^p \\
    \mathcal{H}_T
        &= \sum_{i,\sigma}\Bigg[
            T_{pd}d_{i,\sigma}^\dagger\left(
                p_{i+y/2,\sigma}-p_{i+x/2,\sigma}-p_{i-y/2,\sigma}+p_{i+x/2,\sigma}
            \right)\\&\qquad 
            +T_{pp}\left(
                p_{i+y/2,\sigma}^\dagger p_{i+x/2,\sigma} - p_{i-x/2,\sigma}^\dagger p_{i+y/2,\sigma}
                + p_{i-y/2,\sigma}^\dagger p_{i-x/2,\sigma} - p_{i+x/2,\sigma}^\dagger p_{i-y/2,\sigma}
            \right)
            + \text{h.c.}
        \Bigg].
\end{align*}
$$
ここで、

- $d_{i,\sigma}$: サイト $i$ にある $3d_{x^2-y^2}$ 軌道のスピン $\sigma$ のホールの消滅演算子
- $n_{i,\sigma}^d$: はその数演算子
- $p_{i\pm x(y)/2,\sigma}$: 銅イオンを $x(y)$ 方向につなぐサイト $i\pm x(y)/2$ にある $2p_x(2p_y)$ 軌道のスピン $\sigma$ のホールの消滅演算子
- $n_{i\pm\delta/2,\sigma}^p$: その数演算子。

である。

ホッピングの符号はホールのホッピングになっていることから、
軌道の位相が同じ時には $T>0$, そうでないときには$T<0$である。
