## 銅酸化物様結晶と交代磁性

## 考える系の状況

### ハミルトニアン

交代磁性についての Smejkal-Sinova の Phys.Rev.X 12040501 にある Ru2 の模式図を見ると、
銅酸化物の二次元正方格子を歪ませたように見える。
なので Toy model として銅酸化物をひずませた模型を作って、
交代磁性が出るかを確認してみるというアイデアが出てもおかしくはない。
しかし、遠山先生曰くなぜかやった結果というのが出回っていないらしい。

はじめ単なる模型かと思いきや、層状の結合を忘れれば、CuF2 がその形結晶構造となっている。
なので CuF2 の交代磁性を調べてみたというものになる。

なのでこのモデルを調べて、相図やスピン伝導度、最終的には超伝導との関連も見たいといった目標である。

一般的な銅酸化物をモデルとした模型のハミルトニアンは次のように書ける。
$$
\begin{align*}
    \mathcal{H} 
        &= \mathcal{H}_\varepsilon + \mathcal{H}_U + \mathcal{H}_T\\
    \mathcal{H}_\varepsilon 
        &= \varepsilon_d \sum_{i,\sigma} n_{i,\sigma}^d + (\varepsilon_d +\Delta)\sum_{i,\delta,\sigma}n_{i+\delta/2,\sigma}^p \\
    \mathcal{H}_U 
        &= U_d \sum_i n_{i,\uparrow}^d n_{i,\downarrow}^d 
            + U_p \sum_{i,\delta}n_{i+\delta/2,\uparrow}^p n_{i+\delta/2,\downarrow}^p
            + U_{pd} \sum_{i,\delta,\sigma,\sigma'}n_{i,\sigma}^d n_{i+\delta/2,\sigma'}^p \\
    \mathcal{H}_T
        &= \sum_{i,\sigma}\Bigg[
            T_{pd}d_{i,\sigma}^\dagger\left(
                p_{i+y/2,\sigma}-p_{i+x/2,\sigma}-p_{i-y/2,\sigma}+p_{i+x/2,\sigma}
            \right)\\&\qquad 
            +T_{pp}\left(
                p_{i+y/2,\sigma}^\dagger p_{i+x/2,\sigma} - p_{i-x/2,\sigma}^\dagger p_{i+y/2,\sigma}
                + p_{i-y/2,\sigma}^\dagger p_{i-x/2,\sigma} - p_{i+x/2,\sigma}^\dagger p_{i-y/2,\sigma}
            \right)
            + \text{h.c.}
        \Bigg].
\end{align*}
$$
ここで、

- $d_{i,\sigma}$: サイト $i$ にある $3d_{x^2-y^2}$ 軌道のスピン $\sigma$ のホールの消滅演算子
- $n_{i,\sigma}^d$: はその数演算子
- $p_{i\pm x(y)/2,\sigma}$: 銅イオンを $x(y)$ 方向につなぐサイト $i\pm x(y)/2$ にある $2p_x(2p_y)$ 軌道のスピン $\sigma$ のホールの消滅演算子
- $n_{i\pm\delta/2,\sigma}^p$: その数演算子。

である。

ホッピングの符号はホールのホッピングになっていることから、
軌道の位相が同じ時には $T>0$, そうでないときには$T<0$である。

Slater-Koster table をよりホッピングの値は方向余弦$(l,m,n)$を用いて
$$
\begin{align*}
    T_{pp} &= lmV_{pp\sigma}-lmV_{pp\pi} = \frac{2a(1-a)}{a^2+(1-a)^2}\frac{V_{pp\sigma}-V_{pp\pi}}{2}= \frac{2a(1-a)}{a^2+(1-a)^2}T_{pp0}\\
    T_{pd} &= \frac{\sqrt{3}}{2}l(l^2-m^2)V_{pd\sigma}+l(1-l^2+m^2)V_{pd\pi} = \frac{\sqrt{3}}{2}V_{pd\sigma} = T_{pd0}
\end{align*}
$$
となる。
ここで $a$ は Cu と O の距離であり、これを変えたときの秩序変数を見たいので明記している。
また、$T_{pd}$ の $a$ 依存性を入れてみたいが、思いつかないので同じものであるとする。

A. K. McMahan, Prichar M. Martin, S.Satpathy (PhysRevB 386650 1988) に乗ってる Slater-Koster パラメータはこんな感じ
$$
\begin{align*}
    V_{pp\sigma} &= 1.0 \,\text{eV}\\
    V_{pp\pi} &= -0.3\,\text{eV}\\
    V_{pd\sigma} &= -1.85\,\text{eV}\\
    V_{pd\pi} &= -1.85\,\text{eV}\\
    U_d &= 8.5 \,\text{eV}\\
    U_p &= 4.1\,\text{eV} \sim 7.3 \,\text{eV}\\
    U_{pd} &= 0.6\,\text{eV} \sim 1.3 \,\text{eV}\\
    \Delta = E_p - E_d &= 3.0\,\text{eV}
\end{align*}
$$

また、 Physics of Transition of Metal Oxides の中にあるパラメータだと、
$$
\begin{align*}
    U_d &= 8\\
    U_p &= 4\\
    U_{pd} &= 1\\
    \Delta = E_p-E_d &= 3\\
    T_{pd} &= 1\\
    T_{pp} &= 0.5
\end{align*}
$$
となっている。
また、注意したいのが波動関数の位相によるホッピングの符号の違いがあるというのにも気を付けたい

平均場で使うクーロン相互の形は
$$
    \mathcal{H}_C =  U_d \sum_{k}\frac{1}{2}m(n_{k\downarrow}^d-n_{k\uparrow}^d)
$$

$$
    m = \braket{n_{\uparrow}^d}-\braket{n_{\downarrow}^d}
$$

とりあえず $U_p=0,\,U_{pd}=0$というようにこの相互作用を無視してやってみる。
また、酸素の位置は変わらずホッピングだけが変わらないものとして、
格子定数は元の二倍を基準とする。
パラメータを
$$
\begin{align*}
    U_d &= 8\\
    T_{pp0} &= \frac{V_{pp\sigma}-V_{pp\pi}}{2} =  \frac{1+0.6}{2}\\
    T_{pd0} &=  \frac{\sqrt{3}}{2}V_{pd\sigma} = \frac{\sqrt{3}}{2} \times 1.85\\
    E_p &= 3.0\\
    n_p &= 2.0\\
    T_{pp} &= \frac{2a(1-a)}{a^2+(1-a)^2}T_{pp0}\\
    T_{pd1} &= T_{pd0} \times 1.1 \\
    T_{pd2} &= T_{pd0} \times 0.9 
\end{align*}
$$
とする。ここで$T_{pd1},T_{pd2}$は銅と酸素の距離が2種類あるのでそれぞれに対応するホッピング、
$n_p$ は単位胞内のホールの数で、単位胞内にある2つの銅イオンが持っているとした。
ハミルトニアン行列は

$$
\mathcal{H}(k, a,\pm\sigma) = \begin{pmatrix}
\mp U_d \delta/2 & 0 & T_{pd1}e^{i(-k_x+k_y)a/2} & T_{pd2} e^{i(k_x+k_y)(1-a)/2} & T_{pd1}e^{i(k_x-k_y)a/2} & T_{pd2}e^{-i(k_x+k_y)(1-a)/2}\\
& \pm U_d\delta/2 & T_{pd2}e^{i(k_x-k_y)(1-a)/2} & T_{pd1} e^{-i(k_x+k_y)a/2} & T_{pd2}e^{i(-k_x+k_y)(1-a)/2} & T_{pd1}e^{i(k_x+k_y)a/2}\\
& & E_p & -2T_{pp}\cos\frac{k_x}{2}e^{ik_y(1-2a)/2} & 0 & -2T_{pp}\cos\frac{k_y}{2}e^{-ik_x(1-2a)/2} \\
& & & E_p & -2T_{pp}\cos\frac{k_y}{2}e^{-ik_x(1-2a)/2} & 0\\
& & & & E_p & -2T_{pp}\cos\frac{k_x}{2}e^{-ik_y(1-2a)/2}\\
& & & & &E_p
\end{pmatrix}
$$

```python
def make_delta_graph(x, y, a):
    fig = plt.figure()
    ax1 = fig.add_subplot(111)
    ax1.set_xlim(2,10)
    # ax1.set_yscale("log")
    ax1.set_title("a = {:02d}".format(int(a*100)))
    ax1.set_xlabel("Np")
    ax1.set_ylabel("delta")
    ax1.plot(x, y + 1e-7)

    path = f"../image/xt/"
    if not os.path.isdir(path):
        os.makedirs(path)
    # 出力
    save_folder = "./output/phase/"
    if not os.path.isdir(save_folder):
        os.makedirs(save_folder)
    image_path = save_folder+"a{:02d}.png".format(int(a*100))
    plt.savefig(image_path)

    return image_path
```
