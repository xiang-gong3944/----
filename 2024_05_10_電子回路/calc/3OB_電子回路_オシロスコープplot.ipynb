```python
# prompt: file_data のフォルダにある1.00+E01のようにで書かれた csv ファイルの数値を10.0 のような小数表記に直したうえで 1列目を横軸, 2列目3列目を縦軸にプロットしたグラフをまとめて out_fig のフォルダに出力する

import glob
import pandas as pd
import matplotlib.pyplot as plt

# ファイルパスのリストを取得
file_list = glob.glob(f'./file_data/*.csv')

# グラフを保存するフォルダを作成
!mkdir out_fig

# グラフのインデックス
fig_index = 0

# ファイルごとにグラフを作成
for file_path in file_list:
    # ファイル名から番号を取得
    file_name = file_path.split('/')[-1].split('.')[0]

    # csvファイルの読み込み
    df = pd.read_csv(file_path, header=None)

    # 数値データの読み込み
    data = df.iloc[:, 1:].apply(pd.to_numeric)

    # 数値を10.0のような小数表記に変換
    data = data / 1.00E+01

    # グラフの作成
    plt.figure()
    plt.plot(data.iloc[:, 0], data.iloc[:, 1], label='y1')
    plt.plot(data.iloc[:, 0], data.iloc[:, 2], label='y2')
    plt.legend()
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(file_name)

    # グラフの保存
    save_path = f'./out_fig/{file_name}.png'
    plt.savefig(save_path)
    plt.close()

    # インデックスをインクリメント
    fig_index += 1

```

```python
# prompt: file_data のフォルダにある複数の csv ファイルの 1列目を横軸, 2列目3列目を縦軸にプロットしたグラフをまとめて out_fig のフォルダに出力する

import os
import matplotlib.pyplot as plt
import pandas as pd

# Get the list of CSV files in the file_data folder
csv_files = [f for f in os.listdir('file_data') if f.endswith('.csv')]

# Loop over each CSV file
for csv_file in csv_files:
    # Read the CSV file into a Pandas DataFrame
    df = pd.read_csv(os.path.join('file_data', csv_file))
    df = df.apply(pd.to_numeric, errors='coerce')

    # Extract the data from the DataFrame
    x_data = df.iloc[:, 0]
    y1_data = df.iloc[:, 1]
    y2_data = df.iloc[:, 2]

    # Create the plot
    plt.figure()
    plt.plot(x_data, y1_data, label='y1')
    plt.plot(x_data, y2_data, label='y2')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(csv_file)
    plt.legend()

    # Save the plot to the out_fig folder
    plt.savefig(os.path.join('out_fig', csv_file + '.png'))
    plt.close()
```
