csv_plot フォルダにある複数の csv をまとめてプロットし csv_out フォルダに出力する

```python
import os
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Choose IO folders
input_folder_name = 'csv_plot'
output_folder_name = 'csv_out'

# Get the list of CSV files in the file_data folder
csv_files = [f for f in os.listdir(input_folder_name) if f.endswith('.csv')]

# NMR 特有の設定
x_sup = []
x_inf = []

# Loop over each CSV file
for csv_file in csv_files:
    # Read the CSV file into a Pandas DataFrame
    df = pd.read_csv(os.path.join(input_folder_name, csv_file))
    df = df.apply(pd.to_numeric, errors='coerce')

    # Extract the data from the DataFrame
    x_data = df.iloc[:, 0] *(10**6)
    y1_data = df.iloc[:, 1]
    # y2_data = df.iloc[:, 2]

    #最大値調べる
    x_max = -10
    x_min =  10
    for i in range(x_data.size):
        if x_data[i]>80 : continue
        if x_data[i]<45 : continue
        if x_max < y1_data[i]: x_max=y1_data[i]
        if x_min > y1_data[i]: x_min=y1_data[i]
    x = x_max-x_min
    # Create the plot
    plt.figure()
    plt.plot(x_data, y1_data, label=x)
    # plt.plot(x_data, y2_data, label='y2')
    plt.xlabel('$t$ ($\mu s$)')
    plt.ylabel('$M_z$ ($V$)')
    plt.ylim(-16, 16)
    plt.xlim(0, 400)
    plt.title(csv_file)
    plt.legend()

    # Save the plot to the out_fig folder
    plt.savefig(os.path.join(output_folder_name, csv_file + '.png'))
    plt.close()
```
